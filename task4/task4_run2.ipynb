```python
train_data_path='/content/drive/MyDrive/IberLEF_HAHA_2021/haha_2021_train.csv'
data=pd.read_csv(train_data_path)

del data['is_humor']
del data['votes_no']
del data['votes_1']
del data['votes_2']
del data['votes_3']
del data['votes_4']
del data['votes_5']
del data['humor_rating']
del data['humor_mechanism']

data1=data.dropna(axis=0) #åˆ é™¤targetä¸ºç©ºçš„è¡Œ
```

```python
ls=[]
duan=[]
print(len(data1['humor_target']))
for targ in data1['humor_target']:
  # print(targ)
  targ=targ.replace('; ',';')
  # print(targ)
  ls.append(targ.split(';'))

for ta in ls:
  for t in ta:
    duan.append(t)

print(len(duan))
```

```python
print('writing test result to task2_result')
pd.DataFrame({'task4_humor_target': duan}).to_csv('/content/drive/MyDrive/IberLEF_HAHA_2021/task2/Data/test.csv', index=False)
```
